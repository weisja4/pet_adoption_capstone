```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
import matplotlib.pyplot as plt

rdst = 815
```

```python
pets = pd.read_csv('../data/train/train.csv')
```

```python
pets.head(3)
```

```python
X = pets.drop(columns = ('AdoptionSpeed'))
y = pets['AdoptionSpeed']

X_train, X_hold, y_train, y_hold  = train_test_split(X, y, test_size=0.25, random_state=rdst)

X_train, X_test, y_train, y_test  = train_test_split(X_train, y_train, test_size=0.25, random_state=rdst) # 0.25 x 0.8 = 0.2

print('Pets Shape: ', pets.shape)

print('X Train Shape: ' ,X_train.shape)

print('y Train Shape: ' ,y_train.shape)

print('X Test Shape: ', X_test.shape)
print('y Test Shape: ', y_test.shape)

print('X Holdout Shape: ', X_hold.shape)

print('y Holdout Shape: ', y_hold.shape)

```

```python
hold = pd.concat([X_hold, y_hold], axis = 1)
```

```python
hold.to_csv('../data/holdout.csv', index = False)
```

```python
train = pd.concat([X_train, y_train], axis = 1)
test = pd.concat([X_test, y_test], axis = 1)
pets = pd.concat([train,test], axis = 0)
# train.to_csv('../data/pets')
```
